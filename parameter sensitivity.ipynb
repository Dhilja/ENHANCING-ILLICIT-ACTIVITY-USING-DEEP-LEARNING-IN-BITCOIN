```python
import matplotlib.pyplot as plt
import numpy as np

# File paths for F1 scores for GAT and GIN models
gat_files = ['/content/drive/MyDrive/PROJECT/f1_score_GAT.txt', '/content/drive/MyDrive/PROJECT/f1_scores_GAT_DT.txt', '/content/drive/MyDrive/PROJECT/f1_scores_GAT_DT_comb.txt','/content/drive/MyDrive/PROJECT/f1_scores_GAT_SVM.txt','/content/drive/MyDrive/PROJECT/f1_scores_GAT_SVM_COMB.txt']
gin_files = ['/content/drive/MyDrive/PROJECT/f1_scores_GS.txt', '/content/drive/MyDrive/PROJECT/f1_scores_GS_DT.txt', '/content/drive/MyDrive/PROJECT/f1_scores_GS_DT_COMB.txt','/content/drive/MyDrive/PROJECT/f1_scores_GS_SVM.txt','/content/drive/MyDrive/PROJECT/f1_scores_GS_SVM_COMB.txt']

# Model names
models = ['Model1', 'Model2', 'Model3','Model4','Model5']  # Modify with actual model names

# Load F1 scores for GAT models
f1_scores_gat = [np.loadtxt(file) for file in gat_files]

# Load F1 scores for GIN models
f1_scores_gin = [np.loadtxt(file) for file in gin_files]

# Set the width of the bars
bar_width = 0.35

# Set the position of the bars on the x-axis
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]

# Plotting the bars
plt.bar(r1, f1_scores_gat, color='skyblue', width=bar_width, edgecolor='grey', label='GAT')
plt.bar(r2, f1_scores_gin, color='b', width=bar_width, edgecolor='grey', label='GraphSAGE')

# Adding labels to the bars
plt.xlabel('Models', fontweight='bold')
plt.ylabel('F1 Score', fontweight='bold')
plt.xticks([r + bar_width/2 for r in range(len(models))], models)

# Move the legend to the right and adjust its position
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.yticks(np.arange(0, 1.1, 0.1))

# Showing plot
plt.show()
```
